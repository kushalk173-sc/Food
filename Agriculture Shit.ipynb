```python
import pandas as pd 
df = pd.read_csv('agro.csv')
df
```

```python
df = df.dropna(subset=['WEIGHTED_AVERAGE_PRICE'])
df
```

```python
df = df[df['LEVEL_1'] == 'CHICKEN']
df
```

```python
df.columns
```

```python
df['REPORT_DATE'].unique()
```

```python
df_average_price = df.groupby('REPORT_DATE')['WEIGHTED_AVERAGE_PRICE'].mean().reset_index()
df_average_price
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Sets the figure size
plt.plot(df_average_price['REPORT_DATE'], df_average_price['Average_Price'], marker='o')  # Line plot with markers
plt.title('Average Price Over Time')  # Title of the plot
plt.xlabel('Report Date')  # X-axis label
plt.ylabel('Average Price')  # Y-axis label
plt.grid(True)  # Adds a grid
plt.xticks(rotation=45)  # Rotates x-axis labels to make them more readable
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlap
plt.show()  # Displays the plot
```
